## Line plot

We can make a line plot of one column against the other!

Syntax:

df.plot(x= 'x_values_column', y= 'y_values_column )

````python
# change in glacial volume per year in Europe
df.plot(x='year', y='europe')
````

Customization arguments:

````python
df.plot(x='year', 
        y='europe',
        title='Change in glacial volume per year in Europe',
        xlabel='Year',
        ylabel='Change in glacial volume (km3)',
        color='green'
        )
````

Lets say we want to graph the change in glacial volume in the Arctic and Alaska. We can do it by updating these arguments:

- y: list of column names that will be plotted against x
- color: specify color of each column with a dictionary {'col1' : 'color1', 'col2' : 'color2'}

````python
# dictionary for colors
color = {'arctic':'#F48FB1',
                 'alaska': '#AB47BC'
                 }

df.plot(x='year', 
        y=['arctic', 'alaska'], # list of column names
        title = 'Change in glacial volume per year in Alaska and the Arctic',
        xlabel='Year',
        ylabel='Change in glacial volume (km3)',
        color=color #name of colors dictionary
        )
````

We can also create separate plots for each column by setting the subplots to True

````python
df.plot(x='year', 
        y=['arctic', 'alaska'],
        title = 'Change in glacial volume per year in Alaska and the Arctic',
        xlabel='Year',
        ylabel='Change in glacial volume (km3)',        
        color = color,
        subplots=True
        )
````

# kind() argument in plot

plot() default is a line plot, the parameter that controls the type of plot is kind()

# Scatterplots

Ex: Compare flipper length vs body mass

````python
penguins.plot(kind='scatter',
             x='flipper_length_mm',
             y='body_mass_g',
             alpha=0.4 #controls transparency of points
             )
````

## Bar Plot

We can create bar plots of our data setting kind='bar' in the plot() method.

For example, lets say we want to get data about the 10 penguins with lowest body mass. We can first select this data using the nsmallest() method for series:



````python
# use nsmallest method for series (note nlargest exists, can use for hw)

# 10 smallest penguins
smallest = penguins.body_mass_g.nsmallest(10) #df.column.nsmallest(number of values we want) argument
smallest
````

````python
# let's sort it!
smallest = penguins.body_mass_g.nsmallest(10).sort_values() # in desc order
smallest
````

````python
# let's plot it as a bar chart

smallest.plot(kind='bar')
````

If we wanted to look at other data for these smallest penguins (i.e. what is their species, flipper length, etc) we can use the index of the smallest pandas.Series to select those rows in the original penguins data frame using loc:

````python
penguins.loc[smallest.index] #locate certain rows using loc then use the index of the series to select the rows
````

## Histogram

We can create a histogram of our data setting kind='hist' in plot().

````python
# using plot without subsetting data - a mess again
penguins.plot(kind='hist')
````

To gain actual information, lets subset the data before plotting it. For example, suppose we want to look at the distribution of flipper length. We could do it in this way:

````python
# distribution of flipper length measurements

# first select data, then plot
penguins.flipper_length_mm.plot(kind='hist',
                                title='Penguin flipper lengths',
                                xlabel='Flipper length (mm)',
                                grid=True)
````
